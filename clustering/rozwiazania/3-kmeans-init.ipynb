# k-means - inicjalizacja

Algorytm zależy od początkowej inicjalizacji (funkcja kosztu jest niewypukła). 

ZADANIE: Proszę uruchomić algorytm na danych Iris 100 razy (środkowe współrzędne) z różnymi inicjalizacjami. Zaraportować błąd średniokwadratowy oraz zwizualizować wynik.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data[:, 1:3]  # we only take the first two features.
y = iris.target

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
k=3

SSE_array_random = []
for i in range(100):
    km_random = KMeans(n_clusters=k, n_init=1, init='random', random_state = i)
    y_km_random = km_random.fit_predict(X)
    
    SSE_array_random.append(km_random.inertia_)
    
    plt.scatter(X[:, 0], X[:, 1], c=y_km_random)
    plt.show()
    
```

```python
plt.plot( SSE_array_random, label="Random initialization", alpha = 0.5)
plt.show()
```

Dlatego warto uruchamiać k-means kilka razy i brać wynik z najniższym kosztem

ZADANIE: Proszę ustawić inicjalizację jako k-means++ i porównać

```python
k=3

SSE_array_random = []
SSE_array_pp = []
for i in range(100):
    km_random = KMeans(n_clusters=k, n_init=1, init='random', random_state = i)
    km_pp = KMeans(n_clusters=k, n_init=1, init='k-means++', random_state = i)

    y_km_random = km_random.fit_predict(X)
    y_km_pp = km_pp.fit_predict(X)
    
    SSE_array_random.append(km_random.inertia_)
    SSE_array_pp.append(km_pp.inertia_)
    
#     plt.scatter(X[:, 0], X[:, 1], c=y_km)
#     plt.show()
    
# plt.legend()
# plt.show()
```

```python
plt.plot( SSE_array_random, label="Random initialization", alpha = 0.5)
plt.plot( SSE_array_pp, label="k-means++ initialization", alpha = 1 )
# plt.ylim(40,60)
plt.legend()
plt.show()
```
