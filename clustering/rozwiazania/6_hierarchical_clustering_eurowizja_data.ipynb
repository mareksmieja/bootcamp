```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
dane = loadmat("./eu_song_2014.mat")["x"][0][0][0]
k1 = loadmat("./eu_song_2014.mat")["x"][0][0][1][0][0]
k2 = loadmat("./eu_song_2014.mat")["x"][0][0][5]
dane = pd.DataFrame(dane,columns=k2)
dane.set_index(k1,inplace=True)
```

```python
dane.head()
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
from sklearn import metrics
```

```python
X = dane
df1 = pd.DataFrame(X)
df1

plt.figure(figsize=(25, 10))   
row_clusters = linkage(pdist(df1, metric='euclidean'), method='average') 
row_dendr = dendrogram(row_clusters,
           labels = dane.index, 
           leaf_rotation=45,
           leaf_font_size=15)

plt.ylabel('Euclidean distance')
plt.show()
```

```python
part = fcluster(row_clusters, 6, 'maxclust')
part
```
